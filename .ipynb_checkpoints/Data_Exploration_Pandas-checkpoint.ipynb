```python
dollar = lambda x: float(x)
df.item_price = df.item_price.apply(dollar)
```

```python
#How much was the revenue for the period in the dataset?
revenue = (df['quantity']* df['item_price']).sum()
print(f"Revenue was: ${revenue}")
```

```python
#print item_price per each item_name


filtered = df.drop_duplicates(['item_name', 'item_price'])
one_prod = filtered[filtered.quantity==1]
price_per_item = one_prod[['item_name', 'item_price']]
price_per_item = price_per_item.sort_values('item_price')
print(price_per_item)
```

```python
beer = pd.read_csv('beer.txt', sep=' ')
beer
```

```python
#Which beer has highest cost?
sorted_beer_data = beer.sort_values(by=['cost', 'calories'], ascending=[False, True])
print(sorted_beer_data)
```

```python
#All beers of price==0.40

beer_40=beer[beer.cost==0.43]
beer_40
```

```python
#Sum of calories of those  beers of price==0.43
beer_40=beer[beer.cost==0.43]
print(beer_40['calories'].mean())
print(beer_40['calories'].sum())
```

```python
#average calories and cost for each alcohol category.
beer_groups = beer.groupby('alcohol')
beer_stats = beer_groups[['calories','sodium', 'cost']].mean()
print(beer_stats)
```

```python
crime=pd.read_csv('US_Crime_Rates_1960_2014.csv')
```

```python
crime.head(10)
```

```python
#Convert the type of the column Year to datetime64
crime.Year=pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
#Set the Year column as the index of the dataframe
crime=crime.set_index('Year')
crime.head()
```

```python
#Group the year by decades and sum the values
crimes = crime.resample('10AS').sum()
crimes
```

```python
#Return the first 3 rows of the DataFrame df.
crime.iloc[:3]
```

```python
#Select just the 'Murder' and 'Robbery' columns from the DataFrame
df=crime[['Murder','Robbery']]
df.head()
```

```python
crime.loc[:,['Murder','Robbery']].head()
```

```python
#Select the data in rows [3, 4, 8] and in columns ['Murder', 'Robbery']
crime.loc[df.index[[3, 4, 8]], ['Murder', 'Robbery']]
```

```python
#Select only the rows where the number of murder is greater than 24,000
crime['Murder']>24000
```

```python
crime[crime['Murder']>24000]
```

```python
#Select the rows the murder is between 20k and 24k (inclusive)
filtered_crime = crime[(crime['Murder'] >= 20000) & (crime['Murder'] <= 24000)]
filtered_crime
```

```python
crime[crime['Murder'].between(20000, 24000)]
```

```python
#Calculate the mean murder for each different year
crime.groupby('Year')['Murder'].mean()
```

```python
#Sort crime first by the values in the 'Murder' in decending order,
#then by the value in the 'Violent' column in ascending order.
crime = crime.sort_values('Murder', ascending=False)

crime=crime.sort_values('Violent')

crime.head()
```
